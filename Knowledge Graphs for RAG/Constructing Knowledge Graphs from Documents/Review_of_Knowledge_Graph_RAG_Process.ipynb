# Review of Process for KG --> RAG
* Started wtih a minimum viable graph (MVG), then:
    * Extract - identify interesting information
    * Enhance - "superchage" data (e.g. vector embeddings)
    * Expand - connect information to expand contextual/semantic relationships
    * Repeat process to grow graph
