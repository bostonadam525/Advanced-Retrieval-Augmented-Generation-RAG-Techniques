### Additional Notes about the Documents/Data:
* Completed 10-K forms are available to download as text files that contain XML components --> these are XML files.
* So in order to pre-process the data we will need to perform data cleaning and parsing steps, which the following was done:
* Steps/Process:
1. regex "clean-up" for semantic relevant chunks
2. parsed XML into python data structures using Beautiful Soup web scraping library.
3. Extracted the **CIK (Central Index Key) ID** which is the identifier used by the SEC for each company.
4. For LARGE TEXT CHUNKS, extracted:
  * items 1, 1a, 7 and 7a
5. After data extraction and cleaning, exported to JSON format to create knowledge graph.

# Plan/Process for Constructing a Knowledge Graph from Documents
1. **Chunking** - split form sections into chunks using textsplitter (LangChain).
2. **Nodes** - create graph where each chunk is a node.
  * Metadata of each node added as a property.
3. **Vector Index** - create vector index
4. **Calculate text embedding vectors** for each chunk and populate index
5. **Similarity Search** for relevant chunks
