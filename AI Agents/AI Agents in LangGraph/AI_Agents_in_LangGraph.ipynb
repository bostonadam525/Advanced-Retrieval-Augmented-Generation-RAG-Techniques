# What are Agentic Workflows?
* The basic premisis is an "iterative" workflow.
* A good example:
  * If you were writing a research paper with a group of colleagues, each person would perform a step and you would iterate over the process until complete.

# Why are Agents important?
* The most common way that people interact with LLMs is using a prompt or prompts.
* As an example, zero-shot, one-shot or few-shot prompts produce 1 continuous or complete output.
* Agents in contrast are able to interact with different "steps" along the way to produce a better output.

# Key design patterns of Agentic Workflow
* Planning
* Tools (function calling)
* Reflection (iterative improving results and metrics)
* Multi-process communication
* Memory

# ReAct - Reason, Action

* Giving better results than Chain of Thoughts, ReAct incorporates both reasoning and action at the same time in a prompt. So, while solving a complex problem, ReAct makes sure the LLM performs the below 3 tasks for every subtask

`Reason: Reasoning current states and what’s need to be done`

`Action: Take an action depending on Reason`

`Observation: Results after taking action`

`A ReAct prompt may look something like this (not restricted to this only)`

## Example of ReAct in action:

Solve the given task by following Thought, Action and Observation for every step.
Thought can reason about the current situation and next steps to follow,
Observation is what results are achieved after taking Action.
Action can be of 3 types

Search [item]: Search about item in knowledge base or internet
Follow-up [keyword]: Come up with new follow-up questions based on keyword
Finish [answer]: Give final answer
