### Step 1: Data Ingestion
* Data or Documents are ingested from "external knowledge" source.
* Data can be:
  1. simply structured
  2. structured
  3. unstructured

### Step 2: Indexing
* Each document ingested is split into chunks and embeddings are created.
* Each embeddings with metadata are stored in an index.
* The process is (generally) as follows:
image.png

### 2b. Why do we need indexing in RAG?
* `Quick Retrieval` - expedite process of finding the most relevant information.
* `Enhanced Accuracy` - improves relevance and quality of information that is retrieved.
* `Scalability` - large volumes of data/documents can more easily be handled by the system.


### Step 3: Retrieval
* **Most important piece of RAG pipeline!**
* Goal: retrieve the most relevant chunks with semantic relation to the query.
* Retrieval pipeline: query => embeddings passed to index => `top-K` relevant chunks are extracted and returned.

image.png

### Step 4: Response Synthesis
* `top-k` relevant chunks + `query` are passed to the LLM to **generate** a response.

image.png

### Step 5: Querying
* Querying done through "query engine interface" in RAG pipeline (i.e. llama index or langchain)
* This puts together the `Retrieval` and the `Response Synthesis`in an end-to-end pipeline.
