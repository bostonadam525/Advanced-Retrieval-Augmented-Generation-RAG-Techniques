# Overview of RAG on Tabular Data
* The basic workflow of a Q&A chatbot pipeline with tabular or structure data is what see below.
* The difference from a traditional RAG pipeline is that you DO NOT need to vectorize or embed the structured database tables.

image.png

* The concept is to use the LLM Agent to turn the users request into a structured SQL (or other database language) query and pass that to the database and then the LLM.
* This has been shown to result in more precise outputs and results.

## Treat each row as 1 chunk of the vector database
* Example below is a tabular dataset from kaggle (Titanic Dataset)

* Each row chunk is a string.
* Convert string --> Vector --> vector database
* If dataset has 30 rows, this will be 30 chunks which are 30 vectors in the vector database.
